### 3. Praktik PySpark Lanjutan
- **Latihan 1**: Memanipulasi Data dengan PySpark.
- Tugas 2: Lakukan filter, penghitungan rata-rata, dan pengurutan data menggunakan PySpark.

```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

```python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import avg

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Membuat DataFrame Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Pekerjaan": ["Guru", "Pengusaha", "Karyawan", "Fullsatck developer"],
    "Hobi": ["Memasak", "Menghitung uang", "Lari", "Memancing"],
    "Gender": ["Laki-laki", "Laki-laki", "Prempuan", "Laki-laki"]
}
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame dari Pandas ke PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Join antara dua DataFrame
df_joined = df.join(df_spark_from_pandas, on="Nama", how="inner")

print("Hasil Penggabungan Data PySpark dan Pandas")
df_joined.show()

# Menghitung rata-rata usia
print("Rata-rata Usia")
df_joined.groupBy().agg(avg("Usia")).show()
```

```python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import avg, max, min

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame PySpark
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Membuat DataFrame Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Pekerjaan": ["Guru", "Pengusaha", "Karyawan", "Fullstack Developer"],
    "Hobi": ["Memasak", "Menghitung uang", "Lari", "Memancing"],
    "Gender": ["Laki-laki", "Laki-laki", "Perempuan", "Laki-laki"]
}
df_pandas = pd.DataFrame(data_pandas)

# Konversi DataFrame dari Pandas ke PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Join kedua DataFrame berdasarkan kolom Nama
df_joined = df.join(df_spark_from_pandas, on="Nama", how="inner")

print("Hasil Gabungan Data PySpark dan Pandas:")
df_joined.show()

# Operasi statistik
print("Statistik Usia:")
df_joined.groupBy().agg(
    avg("Usia"),
    max("Usia"),
    min("Usia")
).show()
```

PySpark ke Pnadas

```python
from pyspark.sql import SparkSession
import pandas as pd

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Membuat DataFrame Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Pekerjaan": ["Guru", "Pengusaha", "Karyawan", "Fullsatck developer"],
    "Hobi": ["Memasak", "Menghitung uang", "Lari", "Memancing"],
    "Gender": ["Laki-laki", "Laki-laki", "Prempuan", "Laki-laki"]
}
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

# Join antara dua DataFrame
df_joined = pd.merge(df_pandas, df_pandas_from_spark, on="Nama")

print("Hasil Penggabungan Data PySpark ke Pandas")
print(df_joined)

#operasi statistik
print("\nUsia maksimum:", df_joined["Usia"].max())
print("Usia minimum:", df_joined["Usia"].min())
print("Rata-rata usia:", df_joined["Usia"].mean())
```
